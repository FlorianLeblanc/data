# Clean up rows and columns

```python
# Source: https://stackoverflow.com/questions/19878250/zero-padding-numpy-array
# Build a zero padded array from uneven length lists
def apply_to_zeros(lst, dtype=np.int64):
    inner_max_len = max(map(len, lst))
    result = np.zeros([len(lst), inner_max_len], dtype)
    for i, row in enumerate(lst):
        for j, val in enumerate(row):
            result[i][j] = val
    return result

candidate_rows_array = apply_to_zeros(candidate_rows)
candidate_columns_array = apply_to_zeros(candidate_columns)

number_of_rows = candidate_rows_array.shape[1]
number_of_columns = candidate_columns_array.shape[1]

print number_of_rows
print number_of_columns
```

```python
def get_points(number, array):
    items = []
    for i in xrange(number):
        item = np.median(array[i])
        if item != 0:
            items.append(int(item))
    return items
```

```python
rows = get_points(number_of_rows, candidate_rows_array.transpose(1,0))
columns = get_points(number_of_columns, candidate_columns_array.transpose(1,0))
print rows
print columns
```

```python
def get_array_indexes_new(array):
    consecutive = 45
    index = 0
    array_indexes = []
    find_next_gray = False
    while index < len(array):
        gray = is_gray(array[index])

        if find_next_gray and gray:
            find_next_gray = False

        if find_next_gray and not gray:
            index += 1
            continue        

        subset = array[index:index+consecutive]
        if (subset == subset[0]).all():
            array_indexes.append(index)
            index += consecutive
            find_next_gray = True
        else:
            index += 1
    return np.array(array_indexes)
```
